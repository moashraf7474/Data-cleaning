```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read the file
df=pd.read_excel("/content/Nashville Housing Data for Data Cleaning.xlsx")
```

```python
#Display the first 5 rows
df.head()
```

```python
#Display the number of rows and columns
df.shape
```

```python
#Generate descriptive statistics for numerical columns
df.describe()
```

```python
## Display a concise summary of the DataFrame including column names, data types, non-null counts, and memory usage
df.info()
```

```python
## Calculate the percentage of missing values
missing=df.isnull().sum()
missing_percentage = (missing / len(df)) * 100
missing_percentage
```

```python
## Fill missing values in the 'PropertyAddress' column with the most frequent value (mode)
df['PropertyAddress'].fillna(df['PropertyAddress'].mode()[0],inplace=True)
```

```python
#Drop rows that contain NaN values
df.dropna(axis=1,inplace=True)
```

```python
## Save cleaned dataset
df.to_csv("/content/Nashville Housing Data for Data Cleaning.xlsx", index=False)
```

```python
# Check dataset info again
df.info()

# Verify missing values are handled
df.isnull().sum()
```

```python
# Visualize outliers in SalePrice column
sns.boxplot(x=df['SalePrice'])
plt.show()

# Remove outliers using IQR method
Q1 = df['SalePrice'].quantile(0.25)
Q3 = df['SalePrice'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['SalePrice'] >= Q1 - 1.5 * IQR) & (df['SalePrice'] <= Q3 + 1.5 * IQR)]
```

```python
# Histogram of SalePriceplt.figure(figsize=(8,5))
sns.histplot(df['SalePrice'], bins=50, kde=True)
plt.title("Distribution of Sale Prices")
plt.xlabel("Sale Price")
plt.ylabel("Frequency")
plt.show()
```

```python
## Boxplot of SalePrice
plt.figure(figsize=(6,4))
sns.boxplot(x=df['SalePrice'])
plt.title("Boxplot of Sale Prices")
plt.show()
```

```python
# Average SalePrice by SoldAsVacant
plt.figure(figsize=(6,4))
sns.barplot(x='SoldAsVacant', y='SalePrice', data=df)
plt.title("Average Sale Price by SoldAsVacant")
plt.show()
```
